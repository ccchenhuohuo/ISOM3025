### Q1

Build a weather forecast model with three climate conditions, that is, sunny,  cloudy and rainy.
The transition matrix is as follows:

$$0.6,\;0.2,\; 0.2\\
0.3,\;0.4,\; 0.3\\
0.2,\;0.3,\;0.5$$
representing
$$P(sunny|sunny),\;P(cloudy|sunny),\; P(rainy|sunny)\\
P(sunny|cloudy),\;P(cloudy|cloudy),\; P(rainy|cloudy)\\
P(sunny|rainy),P(cloudy|rainy),\; P(rainy|rainy).
$$

a. Suppose that today is sunny. Simulate weather process for the next 100 days. and plot the results
b. Do a weather forecast for the next 30 days given that today is sunny, respectively. plot your results
c. Compuate the long-run unconditional stationary distribution of the weather, i.e., $P(sunny),\;P(cloudy),\; P(rainy)$.


```python
# define the possible states of the weather conditions:
StatesData = ['Sunny','Cloudy','Rainy']
```

```python
#transition between the weather conditions will be set as follows:
TransitionStates = [['Su-to-Su','Su-to-Cl','Su-to-Ra'],['Cl-to-Su','Cl-to-Cl','Cl-to-Ra'],['Ra-to-Su','Ra-to-Cl','Ra-to-Ra']]
TransitionMatrix = [[0.60,0.2,0.2],[0.3,0.4,0.3],[0.20,0.30,0.50]]
```

```python
WeatherProcess = list()
NumDays = 100
TodayCond = StatesData[0]
print('Weather initial condition =',TodayCond)
```

```python
import numpy as np
for i in range(0, NumDays):
    if TodayCond == 'Sunny':
        TransCondition = np.random.choice(TransitionStates[0],replace=True,p=TransitionMatrix[0])
        if TransCondition == 'Su-to-Su':
            pass
        elif TransCondition == 'Su-to-Cl':
            TodayCond = 'Cloudy'
        else:
            TodayCond = 'Rainy'

    elif TodayCond == 'Cloudy':
        TransCondition = np.random.choice(TransitionStates[1],replace=True,p=TransitionMatrix[1])
        if TransCondition == 'Cl-to-CL':
            pass
        elif TransCondition == 'CL-to-Ra':
            TodayCond = 'Rainy'
        else:
            TodayCond = 'Sunny'
        
    elif TodayCond == 'Rainy':
        TransCondition = np.random.choice(TransitionStates[2],replace=True,p=TransitionMatrix[1])
        if TransCondition == 'Ra-to-Ra':
            pass
        elif TransCondition == 'Ra-to-Cl':
            TodayCond = 'Cloudy'
        else:
            TodayCond = 'Sunny'
            
    WeatherProcess.append(TodayCond)
    print((TodayCond))
```

```python
import matplotlib.pyplot as plt
plt.plot(WeatherProcess)
plt.show()
```

```python
WeatherPred_Su = list()
WeatherPred_Cl = list()
WeatherPred_Ra = list()
NumDays=30
TodayCond = StatesData[0]
print('Weather initial condition =',TodayCond)

Pred_Su = 1
Pred_Cl = 0
Pred_Ra = 0
```

```python
for i in range(0, NumDays):
    Prio_Su = Pred_Su
    Prio_Cl = Pred_Cl
    Prio_Ra = Pred_Ra

    Pred_Su = Prio_Su * TransitionMatrix[0][0] + Prio_Cl * TransitionMatrix[1][0] + Prio_Ra * TransitionMatrix[2][0]
    Pred_Cl = Prio_Su * TransitionMatrix[0][1] + Prio_Cl * TransitionMatrix[1][1] + Prio_Ra * TransitionMatrix[2][1]
    Pred_Ra = Prio_Su * TransitionMatrix[0][2] + Prio_Cl * TransitionMatrix[1][2] + Prio_Ra * TransitionMatrix[2][2]

    WeatherPred_Su.append(Pred_Su)
    WeatherPred_Cl.append(Pred_Cl)
    WeatherPred_Ra.append(Pred_Ra)
    print('Weather forecast for t+', i, ": ", StatesData[0], Pred_Su, StatesData[1], Pred_Cl, StatesData[2], Pred_Ra)
```

```python
plt.plot(WeatherPred_Su)
plt.plot(WeatherPred_Cl)
plt.plot(WeatherPred_Ra)
plt.show()
```

```python
from numpy.linalg import eig
eig_value, eig_vector =eig(np.transpose(TransitionMatrix))
print(eig_value)
print(eig_vector)
```

```python
x=[eig_vector[0,0],eig_vector[1,0],eig_vector[2,0]]/(eig_vector[0,0]+eig_vector[1,0]+eig_vector[2,0])
print(x)
```

```python
plt.plot(WeatherPred_Su)
plt.plot(WeatherPred_Cl)
plt.plot(WeatherPred_Ra)
#help(plt.axhline)
plt.axhline(y=x[0],linestyle="dashed",color="r")
plt.axhline(y=x[1],linestyle="dashed",color="r")
plt.axhline(y=x[2],linestyle="dashed",color="r")
plt.show()
```

```python
P_0=GOOGData.tail(1)
print(P_0)
Recentprice=GOOGData.tail(10)
print(Recentprice)

# SIMULATE 5 days
np.random.seed(4)
n = 5
SQN = 1


mu = mean
Y_0 = np.log(P_0)
sig = std
Multi_path = list()
last_prices = list()
N=10000
plt.figure(figsize=(12,5))
for i in range(N):
    EpsValues = np.random.randn(n)
    Zk = 0
    SBMotion=list()
    for k in range(n):
        Zk =Zk + SQN*EpsValues[k]
        SBMotion.append(Zk)

    simu_Price=list()
    simu_Price.append(np.exp(Y_0))
    for k in range(n):
        simu_Price.append(np.exp(Y_0+SBMotion[k]*sig+mu*k*SQN*SQN))

    plt.plot(simu_Price)
      
    Multi_path.append(simu_Price)
    last_prices.append(simu_Price[n])
plt.show()
```

```python
print(Multi_path[-3:])
print(last_prices[-3:])
```

```python
plt.figure()
plt.hist(np.array(last_prices),bins=40, density=True, histtype='stepfilled',alpha=0.5)
plt.show()
```

```python
simu_ArithRet=np.array(last_prices)/np.array(P_0)-1

VaR_ret=-np.quantile(simu_ArithRet,0.05)
print(VaR_ret)
```

```python
dollar_invest=1000
VaR_dollar=dollar_invest*VaR_ret

# Print the dollar value of Value-at-risk
print(VaR_dollar)
```
