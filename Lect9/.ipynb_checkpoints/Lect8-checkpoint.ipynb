## Value-at-Risk

the main tools for evaluating financial assets. This is because it allows you to monitor
securities and provides a criterion for the construction of portfolios. A risk measure that has been widely used for over two decades is Value-at-Risk
(VaR). The birth of VaR was linked to the growing need for financial institutions to
manage risk and, therefore, be able to measure it. This is due to the increasingly complex
structure of financial markets.

Actually, this measure was not introduced to stem the limits of variance as a risk measure
since an approach to calculate the VaR value starts precisely from the assumptions of
normality. However, to make it easier to understand, let's enclose the overall risk of a
security into a single number or a portfolio of financial assets by adopting a single metric
for different types of risk.

In the financial context, the VaR is an estimate, given a confidence interval, of how high
the losses of a security or portfolio may be in each time horizon. The VaR, therefore,
focuses on the left tail of the distribution of returns, where events with a low probability of
realization are located. Indicating the losses and not the dispersion of the returns around
their expected value makes it a measure closer to the common idea of risk than variance.



$$VaR(\alpha)=-P^{-1}(\alpha),$$
where $P()$ is the distribuition of return (arithmatic return percentage, or in value)

## Simulating Multiple paths

We will simulate a large number paths, say 1000 paths for AMZN. 

```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

ticker = yf.Ticker('AMZN')
AMZN_df = ticker.history(period="10y")


AMZN_df['Close'].plot(title="AMZN's stock price")
AmznData=AMZN_df['Close']

Rt=np.diff(np.log(AmznData))
hatmu=np.array(Rt.mean())
hatsig = np.array(Rt.std())
print(hatmu)
print(hatsig)
```

We will simulate into the future for 5 days, and generate 1000 paths


```python



P_0=AmznData.tail(1)
print(P_0)
Recentprice=AmznData.tail(10)
print(Recentprice)

# SIMULATE 5 days
np.random.seed(4)
n = 5
SQN = 1


mu=hatmu
Y_0=np.log(P_0)
sig=hatsig
Multi_path=list()
last_prices=list()
N=1000
plt.figure(figsize=(12,5))
for i in range(N):
    EpsValues = np.random.randn(n)
    Zk = 0
    SBMotion=list()
    for k in range(n):
        Zk =Zk + SQN*EpsValues[k]
        SBMotion.append(Zk)

    simu_Price=list()
    simu_Price.append(np.exp(Y_0))
    for k in range(n):
        simu_Price.append(np.exp(Y_0+SBMotion[k]*sig+mu*SQN*k))

    plt.plot(simu_Price)
      
    Multi_path.append(simu_Price)
    last_prices.append(simu_Price[n])
plt.show()
```

```python

plt.figure()
plt.hist(np.array(last_prices),bins=40, density=True, histtype='stepfilled',alpha=0.5)
plt.show()
```

We considering invest in 1000 dollar in AMZN today and hold the stock for 5 days. What is the VaR for $\alpha=0.05$ in terms of dollar value? 


We compute the quantile based on the simulated distribution.



```python
simu_ArithRet=np.array(last_prices)/np.array(P_0)-1

VaR_ret=-np.quantile(simu_ArithRet,0.05)
print(VaR_ret)
```

```python
dollar_invest=1000
VaR_dollar=dollar_invest*VaR_ret

# Print the dollar value of Value-at-risk
print(VaR_dollar)
```
